Histograms for each variable in fish

```python
try:
    fig,ax=plt.subplots(3,4,figsize=(15,10))
    fig.delaxes(ax[2,2]) # delete graph 11, 0 based index here
    fig.delaxes(ax[2,3]) # delete graph 12, 0 based index here
    for i in range(ax.shape[0]):
        for j in range(ax.shape[1]):
            sns.histplot(f[f.columns[i*4+j]], ax=ax[i][j])
            fig.axes[3].set_yscale('log') # SO4
            fig.axes[4].set_yscale('log') # NO3
            fig.axes[5].set_yscale('log') # Cl
except IndexError:
    pass
```
